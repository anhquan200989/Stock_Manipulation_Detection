Data Processing

```python
import pandas as pd
from os import listdir, path
```

```python
folder = "Stock data"
dataframes = {}
for filename in listdir(folder):
    if(filename.endswith('.csv')):
        filepath = path.join(folder, filename)
        df = pd.read_csv(filepath)
        df = df.dropna()
        df['Mã'] = filename[:-4]
        dataframes[filename[:-4]] = df


print(dataframes)
```

```python
print(dataframes['UIC'])
```

Use Data

```python
df_train = pd.DataFrame(columns=['Ngày', 'Lần cuối', 'Mở', 'Cao', 'Thấp', 'KL', '% Thay đổi', 'Mã'])
df_val = pd.DataFrame(columns=['Ngày', 'Lần cuối', 'Mở', 'Cao', 'Thấp', 'KL', '% Thay đổi', 'Mã'])
df_test = pd.DataFrame(columns=['Ngày', 'Lần cuối', 'Mở', 'Cao', 'Thấp', 'KL', '% Thay đổi', 'Mã'])
#Chia data
for key, value in dataframes.items():
    perc1 = 0.6
    perc2 = 0.9
    middle = int(len(value) * perc1)
    later = int(len(value) * perc2)
    train_data = value.iloc[0:middle]
    val_data = value.iloc[middle:later]
    test_data = value.iloc[later:len(value)]
    df_train = pd.concat([df_train, train_data], ignore_index=True)
    df_val = pd.concat([df_val, val_data], ignore_index=True)
    df_test = pd.concat([df_test, test_data], ignore_index=True)
print(df_train)
print(df_val)
print(df_test)

#Gộp data
# df_all =  pd.concat(dataframes.values(), ignore_index=True)
# print(df_all)
```
